```python
%matplotlib inline

# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#Make marker larger with list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])

```

```python
#Make marker larger with for loop
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Set the x-axis to a list of strings for each month.
y_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
x_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]


fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = [i * 5 for i in x_axis], label = 'Boston', marker="o", facecolors="red", edgecolors="black", alpha=2.0)
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Set the y limit between 0 and 45.
plt.xlim(0, 50)
# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
plt.title('Pyber Fare by Month')
```
